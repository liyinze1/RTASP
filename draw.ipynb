```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data


y_udp = [690, 1200, 1910, 3230, 3420, 4410]
y_rtasp = [570, 1230, 1630, 2630, 2960, 4170]
y_tcp = [1120, 1700, 2000, 2650, 3050, 3840]

# Create the plot
plt.figure(figsize=(5, 3))  # Set the figure size
plt.plot(y_udp, label='UDP')  # Plot the line and add a label for the legend
plt.plot(y_rtasp, label='RTASP')
plt.plot(y_tcp, label='TCP')
# Add title and labels
plt.xlabel('Packet size')
plt.ylabel('Throughput (KB/s)')
# plt.xscale("log")

x = [128, 256, 512, 1024, 2048, 4096]
plt.xticks(range(6), [str(_) for _ in x])
# Add a legend
plt.legend()

# Show the plot
plt.show()
```
